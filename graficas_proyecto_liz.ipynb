```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

#importacion de las librerias que os pueden ayudar a facilitar el desarrollo de las tablas, asi como la graficacion de las mismas #
```

```python
from posixpath import sep
df = pd.read_csv('https://raw.githubusercontent.com/GaelGarciaSanchez/Proyfinal-liz/refs/heads/main/Student%20Depression%20Dataset.csv', sep =',')
df.head()
df.dtypes
```

```python
df.head()
```

```python
df_depresion = df[df['Depression'] == 1]
#seleccionamos solo los datos de las personas que tengan depresion (osea igual a 1)
gender_counts = df_depresion['Gender'].value_counts()
#contamos el total de personas por genero con depresion

fig = px.bar(
    x=gender_counts.index,
    y=gender_counts.values,
    title='Género con Más Casos de Depresión')
# Crear el gráfico de barras
fig.update_layout(
    xaxis_title='Género',
    yaxis_title='Número de Casos de Depresión',
    template='plotly_white'
)
#le da interactividad

#muestra el grafico
fig.show()
```

```python
import plotly.figure_factory as ff

columns_for_heatmap = [
    'Have_you_ever_had_suicidal_thoughts ?',
    'Family_History_of_Mental_Illness'
]

# Calcular la matriz de correlación
correlation_matrix = df[columns_for_heatmap].corr()

# Crear una matriz de anotaciones personalizada para mostrar "Sí" y "No"
annotations = correlation_matrix.round(2).applymap(
    lambda x: "Sí" if x == 1 else "No" if x == -0.08 else f"{x:.2f}"
).values

# Crear el mapa de calor interactivo
fig = ff.create_annotated_heatmap(
    z=correlation_matrix.values,
    x=list(correlation_matrix.columns),
    y=list(correlation_matrix.index),
    annotation_text=annotations,
    colorscale='Viridis'
)

# Personalizar el diseño
fig.update_layout(
    title='Mapa de Calor de Correlación entre Variables Relacionadas con Salud Mental',
    xaxis_title='Variables',
    yaxis_title='Variables',
    template='plotly_white'
)

# Mostrar el gráfico
fig.show()
```

```python
import plotly.express as px

# Filtrar los datos según el estado de depresión
depressed_students = df[df['Depression'] == 1]
non_depressed_students = df[df['Depression'] == 0]

# Crear el histograma interactivo
fig = px.histogram(
    data_frame=df,
    x="Sleep_Duration",
    color="Depression",
    nbins=20,  # Número de bins ajustable
    title="Distribución de Horas de Sueño según el Estado de Depresión",
    labels={"Depression": "Depresión", "Sleep Duration": "Horas de Sueño"},
    barmode="overlay",  # Superpone los histogramas
    template="plotly_white"
)

# Personalizar las etiquetas para los colores
fig.update_layout(
    xaxis_title="Horas de Sueño",
    yaxis_title="Número de Estudiantes",
    legend_title="Estado de Depresión",
)

# Mostrar el gráfico
fig.show()
```

```python
import plotly.express as px

# Crear gráfico de dispersión
fig = px.scatter(
    data_frame=df,
    x="Financial_Stress",
    y="Depression",
    color="Depression",  # Diferenciar los puntos por el estado de depresión
    title="Relación entre Estrés Financiero y Duración del Sueño según el Estado de Depresión",
    labels={"Financial Stress": "Estrés Financiero", "Sleep duration": "Depresion "},
    template="plotly_white"
)
fig.update_layout(
    xaxis_title="Estrés Financiero (1-5)",
    yaxis_title="Duración del Sueño (en horas)",
    legend_title="Estado de Depresión"
)
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Calcular el porcentaje de estudiantes con depresión por cada grupo de horas de trabajo/estudio
percentage_depression = (
    df.groupby("Work/Study_Hours")["Depression"]
    .mean()  # Media representa el porcentaje de 1s (depresión)
    .reset_index()
)

percentage_depression.columns = ["Work/Study_Hours", "Percentage with Depression"]

# Crear el gráfico de líneas
fig = px.scatter(
    data_frame=percentage_depression,
    x="Work/Study_Hours",
    y="Percentage with Depression",
    title="Relación entre Horas de Trabajo/Estudio y Depresión",
    labels={
        "Work/Study Hours": "Horas de Trabajo/Estudio",
        "Percentage with Depression": "Porcentaje de Depresión"
    },
    template="plotly_white"
)
fig.update_layout(
    xaxis_title="Horas de Trabajo/Estudio",
    yaxis_title="Porcentaje de Estudiantes con Depresión",
    yaxis_tickformat=".0%",
    legend_title="Depresión"
)
fig.show()
```
